```r
num <- seq(10378001:10378100)
num
```

```r
x1 <- round(runif(100, min = 80, max = 100))
x1
```

```r
x2 <- round(rnorm(100, mean = 80, sd = 7))
x2
```

```r
x3 <- round(rnorm(100, mean = 83, sd = 18))
x3
```

```r
x3[which(x3 > 100)] <- 100
x3
```

```r
x = data.frame(num, x1, x2, x3)
```

```r
colMeans(x)
```

```r
rowMeans(x)
```

```r
apply(x, 2, mean)
```

```r
apply(x[c('x1', 'x2', 'x3')], 1, sum)
```
